```python
def binSearch(arr, target):
    left, right = 0, len(arr)-1
    f = -1
    while left <= right:
        mid = left + (right-left)//2
        
        if arr[mid] == target:
            break
        elif arr[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    
    if arr[mid] == target:
        return mid
    else:
        x, y = 0,0
        if mid-1 >= 0:
            x = arr[mid-1]
        if mid+1 <= len(arr)-1:
            y = arr[mid+1]
        if abs(target-arr[mid]) < abs(target-x) and abs(target-arr[mid]) < abs(target-y):
            return mid
        if abs(target-x) < abs(target-y):
            return mid-1
        else:
            return mid+1
```

```python
keep = set()
groups = {}
percent = 1
k = 0

while percent > 0.3 and (len(groups.keys()) > 20 or percent == 1):
    percent -= 0.01
    for index,row in enumerate(optDist.values):
        if index not in keep:
            keep.add(index)
            groups[index] = [index]

            for x, column in enumerate(optDist.values):
                if (x not in keep) and (optDist.values[x,index] >=percent) :
                    keep.add(x)
                    groups[index].append(x)

                if x in groups.keys() and x != index and (optDist.values[x,index] >= (percent-0.05)):
                    for i in groups[x]:
                        groups[index].append(i)
                    groups.pop(x)
        elif index in groups.keys():
            for x, column in enumerate(optDist.values):
                
                if x in groups.keys() and x != index and (optDist.values[x,index] >= (percent-0.05)):
                    for i in groups[x]:
                        groups[index].append(i)
                    groups.pop(x)
            
    k += 1
    print(f"{k} Iterations Completed...\n")
```

```python
trial = [i for i in range(len(arr))]
used = [i for i in arr]
minUserArray = []
depots = []
bestScores = []

count = 0
while len(depots) < 15:

    minCost = float('inf')
    minUser = {}
    best = -1

    
    for index, values in enumerate(trial):
        distance = {}
        for i in trial:
            if dist.values[i, index] in distance.keys():
                newDays = dist.values[i, index]+(random.randint(0,1000)/1000000)
                while newDays in distance.keys():
                    newDays = dist.values[i, index]+(random.randint(0,1000)/100000)
                distance[newDays] = i
            else:
                distance[dist.values[i, index]]= i

        user = {}
        sorted_distance = list(distance.keys())
        sorted_distance.sort()

        cost = 0   
        price = 0
        
        i = 0
        while cost < 20000 and i <= len(trial)-1:
            if (used[distance[sorted_distance[i]]] + cost) < 20000:

                if used[distance[sorted_distance[i]]] > 0:
                    cost += used[distance[sorted_distance[i]]]
                    price += used[distance[sorted_distance[i]]] * sorted_distance[i]
                    user[distance[sorted_distance[i]]] = used[distance[sorted_distance[i]]]

            else:
                poll = 20000 - cost

                if used[distance[sorted_distance[i]]] > 0:
                    cost += poll
                    price += (poll * sorted_distance[i])
                    user[distance[sorted_distance[i]]] = poll
            i+=1
            count += 1
            if count%50000 == 0:
                print(f"{count} Iterations... \n")

        if price < minCost:
            minCost = price
            minUser = user
            best = index

    for i in minUser:
        used[i] -= minUser[i]

    depots.append(best)
    minUserArray.append(minUser)
    bestScores.append(minCost)
    
    print(f"Depot Found {best}\n Price: {minCost}\n")
    
```
