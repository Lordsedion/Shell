```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
```

```python
dataset = pd.read_csv('dataset/Biomass_History.csv')
```

```python
print(len(dataset))
dataset = dataset.drop(columns="Index")
dataset.head()
```

```python
[np.mean(dataset.iloc[i]) for i in range(len(dataset))]
```

```python
df = dataset[0:].drop(columns=['Latitude', 'Longitude', '2017'])
df['mean'] = 2
arr = [np.mean(df.iloc[i]) for i in range(len(df))]
for i in range(len(df)):
    df['mean'][i] = arr[i] 
df
# df['mean'] = [np.mean(dataset.iloc[i]) for i in range(len(dataset))]
```

```python
X = df.values
Y = dataset['2017'].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
rd = RandomForestRegressor(max_depth=100)
```

```python
rd.fit(X, Y)
```

```python
df1 = dataset[0:].drop(columns=['Latitude', 'Longitude', '2010'])
df1['mean'] = 2
arr1 = [np.mean(df1.iloc[i]) for i in range(len(df1))]
for i in range(len(df1)):
    df1['mean'][i] = arr1[i] 
df1
```

```python
X_test = df1.values
```

```python
X_test.shape
```

```python
y_pred = rd.predict(X_test)
```

```python
plt.plot(y_pred[:50], color='r', marker='*')
```

```python
for i in range(len(y_pred)):
    print(f"{i+1} Model Prediction: -> {y_pred[i]}")
```

```python
print(mean_absolute_error(Y_test, y_pred))
print(mean_squared_error(Y_test, y_pred))
```
