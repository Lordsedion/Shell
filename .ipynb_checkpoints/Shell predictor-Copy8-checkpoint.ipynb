```python
X = df.drop(columns=['output', 'temperature']).values[:16926]/100
X_test = df.drop(columns=['output',]).values[16926:]/100

Y = df['output'].values[:16926]

X_train, x_test, Y_train, y_test = X[0:16926], X[16920:], Y[0:16926], Y[16920:]
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

model = tf.keras.Sequential([
    LSTM(units=48, activation='relu', ),
])
    
# model.add(LSTM(units=64, ))
# model.add(LSTM(units=64, return_sequences=True))

# model.add(Dense(units=4))
# model.add(Dense(units=8))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae', 'mse'])
```

```python
model.fit(X_train, Y_train, epochs=25)
```

```python
X_test = df.drop(columns=['output','temperature']).values[16926:]/100
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

y_pred = model.predict(X_test)
```
