```python
arr = np.zeros((2418, 2418))

for i in range(len(arr)):
    for index, value in enumerate(X):
        if abs(index-i) <= 96:
            arr[i][index] = distance_matrix[i][index] * value
```

```python
weighted_mean = {}
```

```python
for i in range(len(arr)):
    weighted_mean[i] = (np.sum(arr[i])*0.001)
    print(f"{i} Weighted Mean: {np.sum(arr[i])*0.001}")
```

```python
sorted_dict_by_values = dict(sorted(weighted_mean.items(), key=lambda item: item[1]))
ordered = list(sorted_dict_by_values.keys())
len(ordered)
```

```python
depotsI = []
start = 0
for i in range(25):
    if start+96 < 2418:
        start += 96
    else:
        start += 2418-start
    depotsI.append(start)
len(depotsI)
```

```python
print(depotsI)
```

```python
start = 0
minList = []
for value in depotsI:
    mini = float('inf')
    c = -1
    for i in range(2418):
        if i <= value and i >= start:
            if weighted_mean[i] < mini:
                mini = weighted_mean[i]
                c = i
            print(f"Ordered index: ->{i}   value:{value} start: {start}   Ordered Values: ->{weighted_mean[i]}")
    minList.append(c)
    start += 96
```

```python
cost_array = []
for i in range(len(minList)):
    cost_array.append(weighted_mean[minList[i]])
    print(f"Index: {minList[i]} ---------   Optimal value cost: {weighted_mean[minList[i]]}")
```
