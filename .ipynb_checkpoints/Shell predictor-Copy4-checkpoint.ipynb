```python
bro = { 'latitude': [], 'longitude':[],'May Rainfall': [],'June Rainfall': [],'July Rainfall': [],
    'August Rainfall': [],'September Rainfall': [],'October Rainfall': [],'max': [],
    'min':[],'mean':[],'std':[],'year': [],'prev year 1':[],'prev year 2':[],'prev year 3':[],'prev year 4':[],
    'prev year 5':[],'prev year 6':[],'prev year 7':[],'output': [],
}

bro['May Rainfall'] = [rain[i][4] for i in range(9) for j in range(len(dataset)) ]
bro['June Rainfall'] = [rain[i][5] for i in range(9) for j in range(len(dataset)) ]
bro['July Rainfall'] = [rain[i][6] for i in range(9) for j in range(len(dataset)) ]
bro['August Rainfall'] = [rain[i][7] for i in range(9) for j in range(len(dataset)) ]
bro['September Rainfall'] = [rain[i][8] for i in range(9) for j in range(len(dataset)) ]
bro['October Rainfall'] = [rain[i][9] for i in range(9) for j in range(len(dataset)) ]

bro['longitude'] = [list(dataset['Longitude'])[j] for i in range(9) for j in range(len(dataset))]
bro['latitude'] = [list(dataset['Latitude'])[j] for i in range(9) for j in range(len(dataset))]
bro['year'].extend([2010+i for i in range(9) for j in range(len(dataset))])

max_array = [None for i in range(len(dataset))]
for i in range(8):
    for j in range(len(dataset)):
        max_array.append(max([dataset[f"{2010+_}"][j] for _ in range(i+1)]))
bro['max'] = max_array

min_array = [None for i in range(len(dataset))]
for i in range(8):
    for j in range(len(dataset)):
        min_array.append(min([dataset[f"{2010+_}"][j] for _ in range(i+1)]))
bro['min'] = min_array

mean_array = [None for i in range(len(dataset))]
for i in range(8):
    for j in range(len(dataset)):
        mean_array.append(np.mean([dataset[f"{2010+_}"][j] for _ in range(i+1)]))
bro['mean'] = mean_array

std_array = [None for i in range(len(dataset))]
for i in range(8):
    for j in range(len(dataset)):
        std_array.append(np.std([dataset[f"{2010+_}"][j] for _ in range(i+1)]))
bro['std'] = std_array

prev1_array = [None for i in range(len(dataset))]
for i in range(8):
    for j in range(len(dataset)):
        prev1_array.append(dataset[f"{2010+i}"][j])
bro['prev year 1'] = prev1_array

prev2_array = [None for i in range(len(dataset)*2)]
for i in range(7):
    for j in range(len(dataset)):
        prev2_array.append(dataset[f"{2010+i}"][j])
bro['prev year 2'] = prev2_array

prev3_array = [None for i in range(len(dataset)*3)]
for i in range(6):
    for j in range(len(dataset)):
        prev3_array.append(dataset[f"{2010+i}"][j])
bro['prev year 3'] = prev3_array

prev4_array = [None for i in range(len(dataset)*4)]
for i in range(5):
    for j in range(len(dataset)):
        prev4_array.append(dataset[f"{2010+i}"][j])
bro['prev year 4'] = prev4_array

prev5_array = [None for i in range(len(dataset)*5)]
for i in range(4):
    for j in range(len(dataset)):
        prev5_array.append(dataset[f"{2010+i}"][j])
bro['prev year 5'] = prev5_array

prev6_array = [None for i in range(len(dataset)*6)]
for i in range(3):
    for j in range(len(dataset)):
        prev6_array.append(dataset[f"{2010+i}"][j])
bro['prev year 6'] = prev6_array

prev7_array = [None for i in range(len(dataset)*7)]
for i in range(2):
    for j in range(len(dataset)):
        prev7_array.append(dataset[f"{2010+i}"][j])
bro['prev year 7'] = prev7_array

out_array = []
for i in range(8):
    for j in range(len(dataset)):
        out_array.append(dataset[f"{2010+i}"][j])
        
for i in range(2418):
    out_array.append(None)
bro['output'] = out_array

df = pd.DataFrame(bro)
df.head()
```

```python
model = tf.keras.Sequential([
    LSTM(units=128, activation='relu', ),
    Dense(16),
#     LSTM(units=48, activation='relu', ),
    Dense(8),
#     LSTM(units=52, activation='relu', ),
    Dense(1)
])
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

model.compile(tf.keras.optimizers.Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y, epochs=100,)
```
