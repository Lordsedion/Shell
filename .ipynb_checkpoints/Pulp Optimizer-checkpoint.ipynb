```python
from pulp import *
import random as rnd

sites = [f"Site_{i}" for i in range(100)]
depots = [f"{i}" for i in range(100)]

supply = {f"Site_{i}":rnd.randint(2,100) for i in range(100)}

refineries = [f"i" for i in range(100)]

demand = {f"{i}":1000 for i in range(100)}
distance = [[rnd.randint(0, 200) for i in range(100)] for i in range(100)]

distance = makeDict([sites, depots], distance, 0)

prob = LpProblem("Depot_Problem", LpMinimize)

Routes = [(w, b) for w in sites for b in depots]

depot_var = LpVariable.dicts("Supply_to_depots",depots, 0, 1, LpInteger)

vars = LpVariable.dicts("Route", (sites, depots), 0, None, LpInteger)

prob += lpSum(vars[w][b] * distance[w][b] for (w,b) in Routes)

for w in sites:
    prob += (
        lpSum(vars[w][b] for b in depots) <= supply[w],
        f"Sites must not give more than they produce{w}"
    )

for b in depots:
    prob += (
        lpSum([vars[w][b] for w in sites]) <= demand[b]*depot_var[b],
        f"Sum of products into depot should be less than its capacity {b}"
    )

prob += lpSum(depot_var[b] for b in depots) <= 6

for w in sites:
    prob += (
    lpSum(vars[w][b] for b in depots) >= 0.8* supply[w],
    f"At least 80% processing {w}"
    )

prob.solve()

for i in vars:
    for each in vars[i]:
        print(f"TP {vars[i][each]} ---  {vars[i][each].varValue}")

print("Supplied Depots:")
for b in depots:
    if depot_var[b].varValue == 1:
        print(b)
```

```python
from pulp import *
import random as rnd

sites = [f"Site_{i}" for i in range(10)]
depots = [f"{i}" for i in range(10)]

a = [511, 546, 250, 497, 519, 124, 495, 573, 510, 132]
supply = {f"Site_{i}":a[i] for i in range(10)}


refineries = [f"{i}" for i in range(10)]

demand = {f"{i}":1000 for i in range(10)}
demand_ref = {f"{i}":2000 for i in range(10)}

distance = [[8, 3, 0, 9, 3, 9, 5, 2, 9, 5],
            [6, 4, 1, 1, 7, 1, 2, 0, 8, 2],
            [4, 10, 9, 6, 3, 5, 0, 9, 2, 3],
            [10, 4, 0, 1, 8, 0, 6, 8, 0, 4],
            [8, 6, 0, 2, 6, 5, 1, 9, 1, 6],
            [7, 3, 5, 4, 4, 0, 4, 7, 0, 9],
            [2, 9, 7, 10, 2, 3, 4, 4, 10, 6],
            [9, 9, 10, 1, 10, 1, 3, 10, 4, 4],
            [4, 1, 5, 1, 5, 4, 2, 1, 3, 3],
            [2, 8, 3, 7, 1, 2, 8, 8, 6, 10]]
print(distance)

distance = makeDict([sites, depots], distance, 0)

prob = LpProblem("Depot_Problem", LpMinimize)

Routes = [(w, b) for w in sites for b in depots]

depot_var = LpVariable.dicts("Supply_to_depots",depots, 0, 1, LpInteger)

vars = LpVariable.dicts("Route", (sites, depots), 0, None, LpInteger)

# Ref
ref_vars = LpVariable.dicts("Route_ref", (depots, refineries), 0, None, LpInteger)
ref_var = LpVariable.dicts("Supply_to_refineries",refineries, 0, 1, LpInteger)
Routes_ref = [(w, b) for w in depots for b in refineries]

prob += (
    lpSum(lpSum(vars[w][b] * distance[w][b] for (w, b) in Routes)*0.001 *
        lpSum(ref_vars[b][k] * distance[b][k] for (b, k) in Routes_ref)*0.001)
)
for w in sites:
    prob += (
        lpSum(vars[w][b] for b in depots) <= supply[w],
        f"Sites must not give more than they produce{w}"
    )

#Refineries
for b in depots:
    prob += (
        lpSum(ref_vars[b][k] for k in refineries) <= (lpSum(vars[w][b] for w in sites)),
        f"Depot must not give more than they produce{b}"
    )

for b in depots:
    prob += (
        lpSum([vars[w][b] for w in sites]) <= demand[b]*depot_var[b],
        f"Sum of products into depot should be less than its capacity {b}"
    )

#Refineries
for b in refineries:
    prob += (
        lpSum([ref_vars[j][b] for j in depots]) <= demand_ref[b]*ref_var[b],
        f"Sum of products into refineries should be less than its capacity {b}"
    )

prob += lpSum(depot_var[b] for b in depots) <= max(1, min(6,(0.8*sum(supply.values())//1000)+1))
prob += lpSum(ref_var[b] for b in refineries) <= max(1, min(4,(0.8*sum(supply.values())//2000)+1))

for w in sites:
    prob += (
    lpSum(vars[w][b] for b in depots) >= 0.8* supply[w],
    f"At least 80% processing {w}"
    )

# Contraint 8: Amount entering into depot equals amount exiting depot
for b in depots:
    if depot_var[b] >= 1:
        prob += (
            lpSum(vars[w][b] for w in sites) <= lpSum(ref_vars[b][k] for k in refineries if ref_var[k] >= 1)
        )
    
# Refineries
hufe = ()
for w in depots:
    for b in refineries:
        if ref_var[b] >= 1:
            hufe += (
            lpSum(ref_vars[w][b] )
            )

prob += lpSum(hufe) >= sum(supply.values())*0.8

prob.solve()

for i in vars:
    for each in vars[i]:
        if vars[i][each].varValue > 0:
            print(f"TP {vars[i][each]} ---  {vars[i][each].varValue}")

print("Supplied Depots:")
for b in depots:
    if depot_var[b].varValue == 1:
        print(b)

print("\nSupplied Refineries:")
for b in refineries:
    if ref_var[b].varValue == 1:
        print(b)

print(f"\n{sum(supply.values())}")
print((sum(supply.values())//1000)+1)
print(sum([(lpSum(vars[w][b].varValue for w in sites)) for b in depots if depot_var[b].varValue >= 1]))
print(sum([(lpSum(ref_vars[w][b].varValue for w in depots)) for b in refineries if ref_var[b].varValue >= 1]))
print([depot_var[i].varValue for i in depot_var if depot_var[i] == 1])
print(supply.values())
```
