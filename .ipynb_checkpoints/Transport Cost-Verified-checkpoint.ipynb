```python
def binSearch(arr11, target):
    left, right = 0, len(arr11)-1
    f = -1
    while left <= right:
        mid = left + (right-left)//2
        
        if arr11[mid] == target:
            f = mid
            break
        elif arr11[mid] > target:
            right = mid-1
        else:
            left = mid +1
    return f
```
