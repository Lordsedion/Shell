```python
bro = {
    'latitude': [],
    'longitude':[],
    'rainfall': [],
    'temperature':[],
    'max': [],
    'mean':[],
    'std':[],
    'year': [],
    'prev year 1':[],
    'prev year 2':[],
    'prev year 3':[],
    'prev year 4':[],
    'prev year 5':[],
    'prev year 6':[],
    'prev year 7':[],
    'output': [],
}

lat = []
for i in range(8):
    for j in range(2418):
        lat.append(list(dataset['Latitude'])[j])
bro['latitude'] = lat

bro['rainfall'].extend([r[i][j] for r in rain for j in range(2418) for i in range(12)])

long = []
for i in range(8):
    for j in range(2418):
        long.append(list(dataset['Longitude'])[j])
bro['longitude'] = long

temp = [40, 41.5, 39, 40, 42, 41, 39.5, 40]
bro['temperature'].extend([temp[i] for i in range(8) for j in range(2418)])

bro['year'].extend([2010+i for i in range(8) for j in range(2418)])

max_array = [None for i in range(2418)]
for i in range(0,7):
    for j in range(2418):
        max_array.append(max([dataset[f"{2010+_}"][j] for _ in range(i+1)]))
bro['max'] = max_array

mean_array = [None for i in range(2418)]
for i in range(0,7):
    for j in range(2418):
        mean_array.append(np.mean([dataset[f"{2010+_}"][j] for _ in range(i+1)]))
bro['mean'] = mean_array

std_array = [None for i in range(2418)]
for i in range(7):
    for j in range(2418):
        std_array.append(np.std([dataset[f"{2010+_}"][j] for _ in range(i+1)]))
bro['std'] = std_array

prev1_array = [None for i in range(2418)]
for i in range(7):
    for j in range(2418):
        prev1_array.append(dataset[f"{2010+i}"][j])
bro['prev year 1'] = prev1_array


prev2_array = [None for i in range(2418*2)]
for i in range(6):
    for j in range(2418):
        prev2_array.append(dataset[f"{2010+i}"][j])
bro['prev year 2'] = prev2_array

prev3_array = [None for i in range(2418*3)]
for i in range(5):
    for j in range(2418):
        prev3_array.append(dataset[f"{2010+i}"][j])
bro['prev year 3'] = prev3_array

prev4_array = [None for i in range(2418*4)]
for i in range(4):
    for j in range(2418):
        prev4_array.append(dataset[f"{2010+i}"][j])
bro['prev year 4'] = prev4_array

prev5_array = [None for i in range(2418*5)]
for i in range(3):
    for j in range(2418):
        prev5_array.append(dataset[f"{2010+i}"][j])
bro['prev year 5'] = prev5_array

prev6_array = [None for i in range(2418*6)]
for i in range(2):
    for j in range(2418):
        prev6_array.append(dataset[f"{2010+i}"][j])
bro['prev year 6'] = prev6_array

prev7_array = [None for i in range(2418*7)]
for i in range(1):
    for j in range(2418):
        prev7_array.append(dataset[f"{2010+i}"][j])
bro['prev year 7'] = prev7_array

out_array = []
for i in range(8):
    for j in range(2418):
        out_array.append(dataset[f"{2010+i}"][j])
bro['output'] = out_array

for i in bro:
    print(i, len(bro[i]))

df = pd.DataFrame(bro)
df.head()
```
