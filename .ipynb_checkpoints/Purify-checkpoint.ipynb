```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        small = []
        big = []
        
        for i in arr[1:]:
            if i< pivot:
                small.append(i)
            else:
                big.append(i)
                
    return quicksort(small) + [pivot] + quicksort(big)
```
